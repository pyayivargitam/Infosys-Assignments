**Matplotlib Library**

matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for data visualization, allowing users to generate plots and graphs from data arrays easily.

Key Features of matplotlib library:
Versatile Plotting: Matplotlib supports a wide variety of plot types, such as line plots, bar charts, scatter plots, histograms, pie charts, 3D plots, heatmaps, and more, making it suitable for nearly all data visualization needs.

Extensive Customization: Almost every element of a plot—colors, markers, line styles, labels, legends, gridlines, fonts, axes, and more—can be customized for maximum control and presentation quality.

Integration with NumPy and Pandas: Data from NumPy arrays or pandas DataFrames can be plotted directly, allowing seamless integration with the broader Python data stack.

Publication-Quality Output: Generate high-quality graphics with fine control over resolution, vector and raster output (PDF, PNG, SVG, etc.), and layout, suitable for academic publications and presentations.

Interactivity: Matplotlib enables dynamic features such as zooming, panning, and updates to figures. It also supports embedding interactive plots within GUIs and Jupyter Notebooks for exploratory analysis.

Subplots and Layouts: Easily create multiple plots in a single figure using subplots, grids, and advanced layout management, helping to compare multiple datasets or perspectives in one visualization

# Line Chart

A line chart (or line graph) is a graphical representation of data points connected by straight lines. It is primarily used to visualize trends, changes, or relationships between two or more variables over a continuous interval, often time. The data points are plotted in order on the x-axis against their corresponding values on the y-axis, allowing viewers to observe how a variable evolves and to identify patterns such as increase, decrease, or volatility.

# Histogram
A histogram is a graphical representation that organizes a group of data points into specified ranges (called bins or intervals). It displays the frequency (count) of data points within each bin as bars, making it useful to visualize the distribution, central tendency, spread, and shape of numerical data.

```python
# Q. How are the prices of different cars distributed?
plt.hist(df['price']) # Histogram of car prices
plt.xlabel('price')
plt.ylabel('number of cars')
plt.title('Distribution of prices of cars');
```

```python
plt.hist(df['price'], bins=5) # Histogram of car prices with 5 bins
plt.xlabel('price')
plt.ylabel('number of cars')
plt.title('Distribution of prices of cars');
```

```python
two_door_df = df[df['doornumber'] == 'two'] # Filter for two-door cars
four_door_df = df[df['doornumber'] == 'four'] # Filter for four-door cars
```

# Subplots
In Matplotlib, subplots allow you to create multiple plots or charts in a single figure, arranged in a grid layout. This is useful for comparing different datasets or showing different views of the same data side by side.

```python
fig, ax = plt.subplots() # Create a figure and axis object
```

Explanation:

plt.subplots() is a function that creates a new figure (fig) and one or more axes (ax) objects.

By default, calling plt.subplots() without any arguments creates:

A new figure—the entire drawing area or canvas where plots are rendered.

A single axes—the area where data is plotted, including x and y axes, labels, titles, and the actual drawing space for graphs.

The function returns two objects:

fig: The Figure object representing the whole window or page that contains everything.

ax: The Axes object, which is the actual plot or graph area inside the figure.

```python
# Q. How are the prices of two-door and four-door cars distributed?
fig, axs = plt.subplots(1, 2, figsize = (8,4)) # Create a figure with 1 row and 2 columns of subplots
axs[0].hist(two_door_df['price'], bins=20, color='blue') # Histogram of two-door car prices
axs[0].set_title('Histogram of prices of two door cars') # Set title for the first subplot

axs[0].set_xlabel('price')
axs[0].set_ylabel('frequency')
axs[0].tick_params(axis='x', rotation=45) # Rotate x-axis labels for better readability

axs[1].hist(four_door_df['price'], bins = 20, color = 'green') # Histogram of four-door car prices
axs[1].set_title("Histograms of prices of four door cars")

axs[1].set_xlabel('price')
axs[1].set_ylabel('frequency')
axs[1].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability

plt.suptitle('Histograms of prices of two door and four door cars') # Set a common title for the entire figure
plt.tight_layout(); # Adjust layout to prevent overlap of titles and labels
```

```python
# Q. How are the prices of two-door and four-door cars distributed?
fig, axs = plt.subplots(1, 2, figsize = (8,4)) # Create a figure with 1 row and 2 columns of subplots
axs[0].hist(two_door_df['price'], bins=20, color='blue') # Histogram of two-door car prices
axs[0].set_title('Histogram of prices of two door cars') # Set title for the first subplot

axs[0].set_xlabel('price')
axs[0].set_ylabel('frequency')
axs[0].tick_params(axis='x', rotation=45) # Rotate x-axis labels for better readability

axs[1].hist(four_door_df['price'], bins = 20, color = 'green') # Histogram of four-door car prices
axs[1].set_title("Histograms of prices of four door cars")

axs[1].set_xlabel('price')
axs[1].set_ylabel('frequency')
axs[1].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability

plt.suptitle('Histograms of prices of two door and four door cars') # Set a common title for the entire figure
plt.tight_layout(); # Adjust layout to prevent overlap of titles and labels
```

```python
#
fig, ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)), doornumber_carbody_crosstable.loc['four'])
bar1 = ax.bar(np.arange(len(carbody_categories)), doornumber_carbody_crosstable.loc['two'])
ax.set_xlabel('Index of Categories')
ax.set_ylabel('Count')
ax.set_title('Distribution of number of cars over number of doors and type of car body');
```

```python
# Grouped bar plot to visualize the distribution of cars based on number of doors and type of car body by controlling the transparency
fig, ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)), doornumber_carbody_crosstable.loc['four'], alpha=0.5)
bar1 = ax.bar(np.arange(len(carbody_categories)), doornumber_carbody_crosstable.loc['two'], alpha=0.5)
ax.set_xlabel('Index of Categories')
ax.set_ylabel('Count')
ax.set_title('Distribution of number of cars over number of doors and type of car body');
```

```python
# Grouped bar plot to visualize the distribution of cars based on number of doors and type of car body by controlling the width of the bars
fig, ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)), doornumber_carbody_crosstable.loc['four'], width=0.5)
bar1 = ax.bar(np.arange(len(carbody_categories)), doornumber_carbody_crosstable.loc['two'],width=0.5)
ax.set_xlabel('Index of Categories')
ax.set_ylabel('Count')
ax.set_title('Distribution of number of cars over number of doors and type of car body');
```

```python
# Grouped bar plot to visualize the distribution of cars based on number of doors and type of car body adjusting the position of the bars
fig, ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories))-0.25, doornumber_carbody_crosstable.loc['four'], width=0.5) # move bars to the left
bar1 = ax.bar(np.arange(len(carbody_categories))+0.25, doornumber_carbody_crosstable.loc['two'], width=0.5) # move bars to the right
ax.set_xlabel('Index of Categories')
ax.set_ylabel('Count')
ax.set_title('Distribution of number of cars over number of doors and type of car body');
```

```python
fig, ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories))-0.2, doornumber_carbody_crosstable.loc['four'], width=0.4,label='four') # move bars to the left and control width
bar1 = ax.bar(np.arange(len(carbody_categories))+0.2, doornumber_carbody_crosstable.loc['two'], width=0.4,label='two') # move bars to the right and control width
ax.set_xlabel('Index of Categories')
ax.set_ylabel('Count')
ax.set_title('Distribution of number of cars over number of doors and type of car body')
ax.legend();
```

The typical steps of data preprocessing in machine learning include the following:

Data Collection: Acquiring the relevant dataset for the problem at hand.

Data Cleaning: Handling missing values, removing or correcting errors, and addressing outliers to ensure data quality.

Data Integration: Combining data from multiple sources if applicable.

Data Transformation: Normalizing, scaling, encoding categorical variables, and other transformations to convert data into a suitable format.

Data Reduction: Reducing dimensionality or selecting important features to simplify the dataset.

Handling Imbalanced Data: Addressing class imbalance if present for classification tasks.

Data Splitting: Dividing the dataset into training, validation, and test subsets for model evaluation.

These steps ensure the dataset is clean, consistent, and well-prepared for modeling, improving the performance and reliability of machine learning models.

**Handling Missing Values:**


*   Many real-world datasets have missing or null values in some features. These missing values can cause errors or degrade the performance of machine learning models if not handled properly.

*   **Imputation:** It is the process of replacing missing values with substituted values based on available data. The goal is to maintain dataset integrity by filling missing entries so the model can learn from a complete dataset.

*   SimpleImputer: This is a scikit-learn utility  used for imputation. It provides different strategies for filling missing values:
*   Mean Imputation: Used here for numerical columns (num_imputer = SimpleImputer(strategy='mean')), it replaces missing numerical values with the mean (average) of the available values in that feature.

*   Most Frequent Imputation: Used here for categorical columns (cat_imputer = SimpleImputer(strategy='most_frequent')), it replaces missing categorical values with the most common (mode) value in the feature.

* Applying Imputers: The fit_transform method computes the statistic (mean or most frequent) from the data and replaces missing values accordingly. This ensures the entire dataset has no missing entries for the selected numerical and categorical columns, making it ready for downstream machine learning tasks.

**1. Label Encoding:**
Converts a binary categorical variable (here, the 'Purchased' column with values like "Yes" or "No") into numeric labels (0 and 1).

Appropriate for ordinal or binary categories.

Uses scikit-learn’s LabelEncoder which assigns numeric codes to each category.

**2. One-Hot Encoding:**
Converts nominal categorical variables (here, 'Gender' and 'City') into multiple binary columns representing each category.

Each original category value gets a separate column with 1 indicating presence, 0 otherwise.

Helps avoid introducing a spurious order that numeric codes might imply.

Uses scikit-learn’s OneHotEncoder. The drop='first' argument drops the first category as a reference to avoid multicollinearity (dummy variable trap).

sparse_output=False ensures the output is a dense array suitable for creating a DataFrame.

```python
# 2. Encoding Categorical Variables

# For 'Purchased' (binary), use Label Encoding
le_purchased = LabelEncoder()
df['Purchased'] = le_purchased.fit_transform(df['Purchased'])

# For nominal categorical features ('Gender', 'City'), use OneHotEncoding

ohe = OneHotEncoder(drop='first', sparse_output=False)
ohe_features = ohe.fit_transform(df[['Gender', 'City']])

# Get column names for one-hot encoded features
ohe_feature_names = ohe.get_feature_names_out(['Gender', 'City'])

# Create DataFrame for encoded features
df_ohe = pd.DataFrame(ohe_features, columns=ohe_feature_names, index=df.index)

# Concatenate encoded columns and drop original ones
df = pd.concat([df.drop(['Gender', 'City'], axis=1), df_ohe], axis=1)
```

3. Scaling Numerical Features (Standardization)
Concept: Feature scaling standardizes numerical data by centering it around the mean and scaling it to unit variance.

StandardScaler:
This scikit-learn class computes the mean and standard deviation of numerical features and transforms each value ( x ) using the formula:

z=(x−μ)/σ

where  μ  is the mean and  σ  is the standard deviation.

Purpose:
Standardized features have zero mean and unit variance, which helps many machine learning algorithms perform better and converge faster, especially algorithms sensitive to feature scale like SVM, logistic regression, and neural networks.

Application:
The numerical columns (num_cols) in the dataset are scaled using fit_transform which fits the scaler on the data and transforms it in a single step.

4. Outlier Detection and Handling (Capping)
Concept: Outliers are extreme values in data that can distort model training and affect results.

Capping by Percentiles (Winsorizing): This technique limits the extreme values by capping them at a specified lower and upper percentile value (e.g., 1st and 99th percentile).

In the code, the variable lower is set as the 1st percentile and upper as the 99th percentile of the Account Balance feature. clip() is then used to replace any values below lower with lower, and above upper with upper. This reduces the influence of extreme outliers while preserving all data points.

Purpose: Helps improve model stability and reduces variance caused by extreme values without completely removing data points.

```python
# 3. Scaling Numerical Features

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

# 4. Outlier Detection and Handling (example: capping by percentiles)

for col in ['Account Balance']:
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    df[col] = df[col].clip(lower, upper)
```

```python
## Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Load and preprocess the dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset/preprocessed_earthquake_data.csv')
drop_columns = ['Type', 'Magnitude Type', 'Status', 'Source', 'Year', 'Day']
X = df.drop(columns=drop_columns + ['Status_Reviewed'])
y = df['Status_Reviewed']
```

```python
# Add the synthetic Gaussian noise to features.
noise = np.random.normal(0, 0.1, X.shape)
X_noisy = X + noise
```

```python
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_noisy, y, test_size=0.2, random_state=42)
```

```python
# Train Logistic Regression with customized parameters
logreg = LogisticRegression(C=0.5, max_iter=200, solver='saga', penalty='l2', random_state=42)
logreg.fit(X_train, y_train)
```

```python
# Train Decision Tree with customized parameters
dtree = DecisionTreeClassifier(max_depth=5, min_samples_split=10, min_samples_leaf=5, random_state=42)
dtree.fit(X_train, y_train)
```

```python
# Model Predictions
y_pred_logreg = logreg.predict(X_test)
y_pred_dtree = dtree.predict(X_test)
```

```python
# Classification metrics and reports
def print_classification_metrics(y_true, y_pred, model_name):
    print(f"--- {model_name} Metrics ---")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("F1-score:", f1_score(y_true, y_pred, average='weighted'))
    print("Classification Report:\n", classification_report(y_true, y_pred))
```

```python
print_classification_metrics(y_test, y_pred_logreg, "Logistic Regression")
print_classification_metrics(y_test, y_pred_dtree, "Decision Tree")
```

```python
# Confusion matrix plotting
def plot_confusion(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
plot_confusion(y_test, y_pred_logreg, "Logistic Regression Confusion Matrix")
plot_confusion(y_test, y_pred_dtree, "Decision Tree Confusion Matrix")
```

```python
# Regression metrics for analysis for Logistic Regression
print("MAE (Logistic Regression):", mean_absolute_error(y_test, y_pred_logreg))
print("MSE (Logistic Regression):", mean_squared_error(y_test, y_pred_logreg))
print("R² score (Logistic Regression):", r2_score(y_test, y_pred_logreg))
```

```python
# Regression metrics for analysis for Decision Tree
print("MAE (Decision Tree):", mean_absolute_error(y_test, y_pred_dtree))
print("MSE (Decision Tree):", mean_squared_error(y_test, y_pred_dtree))
print("R² score (Decision Tree):", r2_score(y_test, y_pred_dtree))
```
